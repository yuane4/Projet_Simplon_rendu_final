```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_excel('Sources.xlsx', encoding='latin-1')
```

```python
df['Adresses'] = df['Adresses'].str.lower()
```

```python
df_sources = sklearn.utils.shuffle(df) #on secoue l'échantillon afin de réorganiser aléatoirement l'échantillon
```

```python
df_sources
```

```python
df_sources.to_excel("shuffle.xlsx")
```

```python
df_sources['Adresses'].dtypes #le type des données est O soit un mix de caractère, string, et chiffre
```

```python
from sklearn.model_selection import train_test_split 
#on divise l'échantillon en train et test 
X = df_sources['Adresses']
y = df_sources['Target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# X_train
```

### modèle Naives Bayes en Multinomial

```python
from sklearn.feature_extraction.text import CountVectorizer 
#on vectorize les données a étudier afin d'être comprise par le modèle
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train.astype('U'))
X_test_counts = count_vect.transform(X_test.astype('U'))
X_train_counts.shape, X_test_counts.shape
```

```python
from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB()
clf.fit(X_train_counts, y_train)

y_pred = clf.predict(X_train_counts)
      
# gnb = GaussianNB()
# y_pred = gnb.fit(X_train, y_train).predict(y_test)
# print("Number of mislabeled points out of a total %d points : %d" 
#        % (X_test.shape[0], (y_train != y_pred).sum()))
```

```python
from sklearn import metrics #on imprime les métrics d'évaluations
print(metrics.classification_report(y_train, y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB

y_pred = clf.predict(X_test_counts) 
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, y_pred))
```

```python
metrics.confusion_matrix(y_test, y_pred)
```

### On change de dataset avec une répartion 99% adresses 1% de bruit

```python
df1 = pd.read_excel('99-1.xlsx', encoding='latin-1')
# df_991
```

```python
df_991 = sklearn.utils.shuffle(df1)
```

```python
df_991.to_excel('df_991.xlsx')
```

```python
df_991.head(5)
```

```python
# len(df_991)
```

```python
from sklearn.model_selection import train_test_split 
#on divise l'échantillon en train et test 
X = df_991['Adresses']
y = df_991['Target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer #on vectorize les données a étudier afin d'être comprise par le modèle
count_vect = CountVectorizer()
X_tr991_counts = count_vect.fit_transform(X_train.astype('U'))
X_te991_counts = count_vect.transform(X_test.astype('U'))
X_tr991_counts.shape, X_te991_counts.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer #afin d'afiner le modèle, on instaure un système de poids probabilitste en Tf -IDF

tf_transformer = TfidfTransformer(use_idf=False).fit(X_tr991_counts)
X_tr991_tf = tf_transformer.transform(X_tr991_counts)
X_te991_tf = tf_transformer.transform(X_te991_counts)
X_tr991_tf.shape, X_te991_tf.shape
```

```python
from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB(alpha =  0.00001, fit_prior = True) 
# clf = MultinomialNB()
clf.fit(X_tr991_tf, y_train)

y_pred = clf.predict(X_te991_tf)
```

```python
from sklearn import metrics #on imprime les métrics d'évaluations
print(metrics.classification_report(y_test, y_pred))
```

```python
metrics.confusion_matrix(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score

X = count_vect.fit_transform(df_991['Adresses'].astype('U'))
y = df_991['Target']

scores = cross_val_score(MultinomialNB(), X, y, cv=8)
print("Cross-validation scores: {}".format(scores))
```

```python
from sklearn.model_selection import GridSearchCV
import time
#Le gridsearch permet de tester les hyperparamètres en même temps dans plusieurs config afin d'en retiré le meilleur paramétrage

parameters = {'alpha': (1, 0.1, 0.01, 0.001, 0.0001, 0.00001), 'fit_prior' : (True, False)}
gs_clf = GridSearchCV(MultinomialNB(), parameters, cv=5, n_jobs=-1)
gs_clf.fit(X, y)  
#print("done in {0}s".format(time() - t0))  
print("Best score: {0}".format(gs_clf.best_score_))  
print("Best parameters set:")  
best_parameters = gs_clf.best_estimator_.get_params()
for param_name in sorted(list(parameters.keys())):  
            print("\t{0}: {1}".format(param_name, best_parameters[param_name]))
```
